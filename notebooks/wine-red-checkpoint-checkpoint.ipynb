### normalizing alcohol and volatile acidity.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['alcohol_scaled', 'volatile_scaled']] = scaler.fit_transform(df[['alcohol', 'volatile acidity']])
df['sulphates_scaled'] = scaler.fit_transform(df[['sulphates']])
df['citric_acid_scaled'] = scaler.fit_transform(df[['citric acid']])
```

## To get more strong corelation i have did a '-' bet 2 strong features ((+ve)alcohol,(-ve coor)volatile acidity)

```python
df['alc_minus_volatile_scaled'] = df['alcohol_scaled'] - df['volatile_scaled']
#checking coorelation between the new derived input and output
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
